```python
import tkinter as tk
from vector import Vector

# ----- Setup Tkinter -----
WIDTH, HEIGHT = 400, 400
root = tk.Tk()
root.title("JUDUL ATAS")
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

# Label info
info_label = tk.Label(root, text="TEXT DIBAWAH", font=("Consolas", 12))
info_label.pack()

# ----- Loop utama -----
def draw():
    canvas.delete("all")

    root.after(50, draw)

draw()
root.mainloop()
```
