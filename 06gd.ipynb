
### Gradient Descent adalah metode untuk mencari nilai parameter (seperti w dan b) yang meminimalkan error (selisih antara prediksi dan data asli). Dalam SGD, kita update parameter setiap data point, bukan setelah semua data dihitung

m += error * x * learning_rate

b += error * learning_rate

ðŸ§  Konsep Stochastic Gradient Descent (SGD)
SGD adalah varian dari Gradient Descent (GD) yang memperbarui parameter model (seperti bobot dan bias) berdasarkan satu data point saja dalam satu iterasi, bukan seluruh dataset.

âœ¨ Perbedaan utama:

Batch GD	Seluruh dataset	Stabil, akurat	Lambat untuk dataset besar

SGD	Satu data point	Cepat, cocok untuk online learning	Lebih noisy, bisa fluktuatif

Mini-batch GD	Kelompok kecil data	Trade-off antara kecepatan dan stabilitas	Butuh tuning ukuran batch


```python
# Bentuk Umum 
# SGD: update untuk setiap data point
for i in range(len(X)):
    # Prediksi untuk satu data point : 
    prediction = np.dot(X[i], w) + b  # gunakan X[i]

    # Hitung error
    error = prediction - y[i]

    # Hitung gradien
    grad_w = error * X[i]  # turunan terhadap w
    grad_b = error         # turunan terhadap b

    # Update parameter
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
```

```python
#Gradient Descent untuk memprediksi persamaan y = mx + b: Contoh Sederhana: Prediksi y = 2x + 1
import numpy as np
import matplotlib.pyplot as plt

# Data sederhana: y = 2x + 1 + noise kecil
X_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([3.1, 4.8, 7.2, 8.9, 11.1])  # mendekati y = 2x + 1

print("Data Training:")
for i in range(len(X_train)):
    print(f"x = {X_train[i]}, y = {y_train[i]}")
```

```python
# Inisialisasi parameter
m = 0.0  # slope awal
b = 0.0  # intercept awal
learning_rate = 0.01
epochs = 100

print(f"\nInisialisasi: m = {m}, b = {b}")

# Simpan history untuk plotting
m_history = []
b_history = []
loss_history = []

# Gradient Descent
for epoch in range(epochs):
    # Hitung prediksi
    y_pred = m * X_train + b
    
    # Hitung error
    error = y_pred - y_train
    
    # Hitung loss (MSE)
    mse = np.mean(error**2)
    loss_history.append(mse)
    
    # Hitung gradien
    # dL/dm = (2/n) * Î£(x * (y_pred - y_true))
    # dL/db = (2/n) * Î£(y_pred - y_true)
    dm = (2/len(X_train)) * np.sum(X_train * error)
    db = (2/len(X_train)) * np.sum(error)
    
    # Update parameter
    m -= learning_rate * dm
    b -= learning_rate * db
    
    # Simpan history
    m_history.append(m)
    b_history.append(b)
    
    # Print setiap 10 epoch
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: m = {m:.3f}, b = {b:.3f}, MSE = {mse:.3f}")

print(f"\nParameter akhir: m = {m:.3f}, b = {b:.3f}")
print(f"Seharusnya mendekati: m = 2.0, b = 1.0")
```

```python
# Plot 1: Loss function
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(loss_history)
plt.title('Penurunan Loss Function')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.grid(True)

# Plot 2: Perubahan parameter m dan b
plt.subplot(1, 3, 2)
plt.plot(m_history, label='m (slope)')
plt.plot(b_history, label='b (intercept)')
plt.title('Perubahan Parameter')
plt.xlabel('Epoch')
plt.ylabel('Nilai Parameter')
plt.legend()
plt.grid(True)

# Plot 3: Garis regresi final
plt.subplot(1, 3, 3)
plt.scatter(X_train, y_train, color='blue', label='Data aktual')
x_line = np.linspace(0, 6, 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', label=f'y = {m:.2f}x + {b:.2f}')
plt.title('Garis Regresi Final')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Penjelasan Matematika Sederhana
# Loss Function (MSE):

# MSE = (1/n) * Î£(y_pred - y_true)Â²
#     = (1/n) * Î£((mx + b) - y_true)Â²

# Gradien:

# dMSE/dm = (2/n) * Î£(x * ((mx + b) - y_true))
# dMSE/db = (2/n) * Î£((mx + b) - y_true)

# Update Rule:
# m = m - learning_rate * dMSE/dm
# b = b - learning_rate * dMSE/db

# Output Contoh:

# Inisialisasi: m = 0.0, b = 0.0
# Epoch 0: m = 0.396, b = 0.140, MSE = 54.814
# Epoch 10: m = 1.672, b = 0.667, MSE = 1.235
# Epoch 20: m = 1.926, b = 0.852, MSE = 0.194
# Epoch 30: m = 1.981, b = 0.922, MSE = 0.143
# ...
# Parameter akhir: m = 2.012, b = 0.974
# Seharusnya mendekati: m = 2.0, b = 1.0
# Kesimpulan: Gradient Descent berhasil menemukan parameter m â‰ˆ 2.0 dan b â‰ˆ 1.0 yang mendekati persamaan asli y = 2x + 1!
```

```python
import numpy as np

# Data sederhana: y = 2x
X = np.array([[1.0], [2.0], [3.0], [4.0]])
y = np.array([2.0, 4.0, 6.0, 8.0])

# Inisialisasi parameter
w = np.array([0.0])  # bobot awal
b = 0.0              # bias awal
learning_rate = 0.01
epochs = 20           # jumlah pengulangan seluruh data
# Untuk menyimpan error per epoch
loss_history = []

# Training dengan SGD
for epoch in range(epochs):
    print(f"=== Epoch {epoch+1} ===")
    total_loss = 0
    for i in range(len(X)):
        # Prediksi untuk satu data point
        prediction = np.dot(X[i], w) + b

        # Hitung error
        error = prediction - y[i]

        # Hitung gradien
        grad_w = error * X[i]
        grad_b = error

        # Update parameter
        w -= learning_rate * grad_w
        b -= learning_rate * grad_b

        # Cetak hasil tiap langkah
        print(f"Data ke-{i+1}: x={X[i][0]}, y={y[i]}")
        print(f"  Prediksi: {prediction:.4f}, Error: {error:.4f}")
        print(f"  Gradien w: {grad_w[0]:.4f}, Gradien b: {grad_b:.4f}")
        print(f"  Update w: {w[0]:.4f}, Update b: {b:.4f}\n")

        total_loss += error**2

    # Rata-rata error per epoch
    mse = total_loss / len(X)
    loss_history.append(mse)
```

```python
import numpy as np

# 1. Prediksi untuk satu data point
def predict(X_i, w, b):
    return np.dot(X_i, w) + b

# 2. Hitung gradien dari satu data point
def compute_gradients(X_i, y_i, w, b):
    prediction = predict(X_i, w, b)
    error = prediction - y_i
    grad_w = error * X_i
    grad_b = error
    return grad_w, grad_b

# 3. Update parameter
def update_parameters(w, b, grad_w, grad_b, lr):
    w -= lr * grad_w
    b -= lr * grad_b
    return w, b

# 4. Fungsi utama SGD
def sgd(X, y, w, b, lr=0.01, epochs=10):
    for epoch in range(epochs):
        for i in range(len(X)):
            grad_w, grad_b = compute_gradients(X[i], y[i], w, b)
            w, b = update_parameters(w, b, grad_w, grad_b, lr)
        print(f"Epoch {epoch+1}: w = {w}, b = {b}")
    return w, b

# 5. Contoh data
X = np.array([[1.0], [2.0], [3.0], [4.0]])  # Fitur
y = np.array([2.0, 4.0, 6.0, 8.0])         # Target

# Inisialisasi parameter
w = np.array([0.0])  # bobot awal
b = 0.0              # bias awal

# Eksekusi SGD
final_w, final_b = sgd(X, y, w, b, lr=0.01, epochs=20)

print(f"\nFinal model: y = {final_w[0]:.2f} * x + {final_b:.2f}")

x_test = np.array([5.0])
y_pred = np.dot(x_test, final_w[0]) + final_b
print(f"Prediksi untuk x=5.0 adalah y={y_pred}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data harga rumah
X_train = np.array([
    [60, 2],
    [80, 3],
    [100, 3],
    [120, 4],
    [140, 4],
    [160, 5],
    [180, 5]
])
y_train = np.array([300, 450, 500, 600, 750, 800, 950])

# Normalisasi fitur
def min_max_normalize(X):
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    return (X - X_min) / (X_max - X_min), X_min, X_max

X_train_norm, X_min, X_max = min_max_normalize(X_train)

# Untuk menyimpan error per epoch
loss_history = []

def train_sgd(X, y, learning_rate=0.01, epochs=100):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0.0

    for epoch in range(epochs):
        total_loss = 0
        for i in range(n_samples):
            prediction = np.dot(X[i], w) + b
            error = prediction - y[i]
            grad_w = error * X[i]
            grad_b = error
            w -= learning_rate * grad_w
            b -= learning_rate * grad_b
            total_loss += error**2
        mse = total_loss / n_samples
        loss_history.append(mse)
    return w, b

def predict(X, w, b):
    return np.dot(X, w) + b

# Training
w, b = train_sgd(X_train_norm, y_train, learning_rate=0.01, epochs=100)

# Plot grafik error
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(loss_history)+1), loss_history, color='purple')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Grafik Error Selama Training SGD (Harga Rumah, Fitur Dinormalisasi)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

ðŸ§  Konsep Batch Gradient Descent (Matrix Form)
Model linier:
y_pred = X_aug @ w
Loss function (MSE):
error = y_pred - y_train
Gradien:
	grad = (2 / len(X_aug)) * (X_aug.T @ error)


1. Konsep Dasar
Ini adalah implementasi Regresi Linear Multiple menggunakan Gradient Descent untuk memprediksi harga rumah berdasarkan dua fitur:

Luas rumah (mÂ²)

Jumlah kamar

2. Persiapan Data

X_train = [[60, 2], [80, 3], [100, 3], ...]  # Fitur: [luas, kamar]

y_train = [300, 450, 500, ...]  # Target: harga (juta)

3. Normalisasi Data

def min_max_normalize(X):

    return (X - X_min) / (X_max - X_min)

Tujuan: Mengubah skala fitur menjadi 0-1 agar training lebih stabil

4. Model Matematika
Rumus Prediksi:

y_pred = w1*x1 + w2*x2 + b

atau dalam bentuk matrix:

y_pred = X @ w

dimana:

w1, w2 = bobot untuk luas dan kamar

b = bias

X = matrix fitur + kolom bias

5. Gradient Descent

Loss Function (MSE):

mse = np.mean((y_pred - y_train)**2)

Rumus: MSE = (1/n) * Î£(prediksi - aktual)Â²

Gradien:

grad = (2 / n) * (X.T @ error)

Turunan MSE terhadap parameter w dan b

Update Parameter:

w -= learning_rate * grad

Mengupdate parameter searah penurunan gradien

6. Proses Training

Alur:

Hitung prediksi: X @ w

Hitung error: prediksi - aktual

Hitung gradien

Update parameter

Ulangi hingga konvergen

7. Visualisasi & Prediksi

Plot loss: Menunjukkan penurunan error selama training

Prediksi baru: Menggunakan parameter yang sudah ditraining

Inti Matematika:


y_pred = w1*luas_norm + w2*kamar_norm + b

Program mencari nilai w1, w2, b yang meminimalkan error prediksi melalui iterasi gradient descent.

Output akhir: Model bisa memprediksi harga rumah baru berdasarkan luas dan jumlah kamar!

```python
import numpy as np
import matplotlib.pyplot as plt

# Data harga rumah
X_train = np.array([
    [60, 2],
    [80, 3],
    [100, 3],
    [120, 4],
    [140, 4],
    [160, 5],
    [180, 5]
])
print(f"X_train:\n {X_train}")
y_train = np.array([300, 450, 500, 600, 750, 800, 950])
print(f"y_train:\n {y_train}")

# Normalisasi fitur
def min_max_normalize(X):
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    return (X - X_min) / (X_max - X_min), X_min, X_max

X_norm, X_min, X_max = min_max_normalize(X_train)
print(f"X_norm:\n {np.round(X_norm, 3)}")

# Tambahkan kolom bias (1) ke X
X_aug = np.hstack([X_norm, np.ones((X_norm.shape[0], 1))])  # shape: (n_samples, n_features+1)
print(f"X_aug:\n {np.round(X_aug, 3)}")

# Inisialisasi parameter
w = np.zeros(X_aug.shape[1])  # termasuk bias
learning_rate = 0.01
epochs = 1000
loss_history = []

# Training dengan Gradient Descent Matrix
for epoch in range(epochs):
    # Model linier: y_pred = X @ w
    y_pred = X_aug @ w

    # Error: selisih prediksi dan target
    error = y_pred - y_train

    # Loss function (MSE): rata-rata error kuadrat
    mse = np.mean(error**2)
    loss_history.append(mse)

    # Gradien terhadap w dan b (gabung jadi satu vektor)
    grad = (2 / len(X_aug)) * (X_aug.T @ error)

    # Update parameter
    w -= learning_rate * grad

# Cetak hasil akhir
print("Bobot akhir:", w[:-1])
print("Bias akhir:", w[-1])

# Plot grafik error
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs + 1), loss_history, color='orange')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Error Training dengan Gradient Descent Matrix')
plt.grid(True)
plt.tight_layout()
plt.show()

# Prediksi rumah baru
X_test = np.array([
    [90, 3],
    [150, 4],
    [200, 5]
])
X_test_norm = (X_test - X_min) / (X_max - X_min)
X_test_aug = np.hstack([X_test_norm, np.ones((X_test_norm.shape[0], 1))])
y_test_pred = X_test_aug @ w

for i, harga in enumerate(y_test_pred):
    print(f"Prediksi rumah {i+1}: {harga:.2f} juta")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Contoh menggunakan data harga rumah (Contoh 1)
X_train = np.array([
    [100, 3, 5, 2],
    [150, 4, 10, 5],
    [200, 5, 2, 1],
    [120, 3, 8, 8],
    [180, 4, 3, 3]
])
y_train = np.array([500, 400, 800, 350, 650])

# Normalisasi fitur
def min_max_normalize(X):
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    return (X - X_min) / (X_max - X_min), X_min, X_max

X_norm, X_min, X_max = min_max_normalize(X_train)

# Tambahkan kolom bias
X_aug = np.hstack([X_norm, np.ones((X_norm.shape[0], 1))])

# Inisialisasi parameter (sekarang 5 parameter: w1,w2,w3,w4,b)
w = np.zeros(X_aug.shape[1])
learning_rate = 0.1
epochs = 1000

# Training
for epoch in range(epochs):
    y_pred = X_aug @ w
    error = y_pred - y_train
    mse = np.mean(error**2)
    grad = (2 / len(X_aug)) * (X_aug.T @ error)
    w -= learning_rate * grad

print("Bobot akhir:", w[:-1])
print("Bias akhir:", w[-1])

# Prediksi
X_test = np.array([[130, 3, 6, 4]])  # dari soal
X_test_norm = (X_test - X_min) / (X_max - X_min)
X_test_aug = np.hstack([X_test_norm, np.ones((X_test_norm.shape[0], 1))])
y_pred = X_test_aug @ w

print(f"Prediksi harga: {y_pred[0]:.2f} juta")
```
