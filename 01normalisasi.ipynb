### normalisasi‚Äìdenormalisasi
________________________________________
üéí Analogi Sehari-hari
Bayangkan kamu punya penggaris:

-	Panjang asli tanah = 50 m¬≤ sampai 150 m¬≤ ‚Üí seperti garis dari 50 cm sampai 150 cm.
-	Normalisasi = kita ‚Äúmenyusutkan‚Äù penggaris itu supaya panjangnya selalu 0 sampai 1.
-	Jadi angka 50 m¬≤ ditempatkan di titik 0, dan 150 m¬≤ ditempatkan di titik 1.

Semua nilai lain otomatis jatuh di antara 0‚Äì1 sesuai posisinya.
- üëâ Normalisasi = mengubah skala tanpa mengubah urutan.
- üëâ Denormalisasi = mengembalikan ke ukuran asli.

xnorm = (x - xmin) / (xmax-xmin)

```python
#contoh 1 konversi nilai [50, 100, 200, 300] ke 0 -1
import numpy as np
# Normalisasi
x = np.array([50, 100, 200, 300]) 
x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
print (x_norm)
#[0.  0.2 0.6 1. ]
#jadi 50 = 0, 100 = 0.2, 300 = 1
```

```python
#contoh 2 konversi nilai [ 50 150] ke 0 -1
x = np.array([50, 150])   # luas tanah
print(x)#[ 50 150]

x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
print (x_norm) #[0. 1.]
```

```python
#contoh 3 konversi nilai [300, 800] ke 0 -1
y = np.array([300, 800])  # harga rumah
y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))
print (y_norm) #[0. 1.]
```

### Denormalisasi : yaitu proses mengubah skala 1 - 0 kembali ke data semula
y = (ynorm * (ymax-ymin)) + ymin

```python
#contoh 4 denormalisasi nilai 0.75
y_pred_norm = 0.75
y_pred = y_pred_norm * (np.max(y) - np.min(y)) + np.min(y)
print(f"{y_pred:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data asli
x = np.array([50, 100, 200, 300])

# Normalisasi Min‚ÄìMax
x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))

# Contoh nilai normalisasi yang ingin dikembalikan
val_norm = 0.5
val_denorm = (val_norm * (np.max(x) - np.min(x))) + np.min(x)

# Buat figure dengan 3 subplot
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Grafik batang data asli
axs[0].bar(range(len(x)), x, color='skyblue')
axs[0].set_title("Data Asli")
axs[0].set_xlabel("Index")
axs[0].set_ylabel("Nilai")
axs[0].set_xticks(range(len(x)))
axs[0].set_xticklabels([f"x{i}" for i in range(len(x))])

# Grafik batang data normalisasi
axs[1].bar(range(len(x_norm)), x_norm, color='orange')
axs[1].axhline(val_norm, color='red', linestyle='--', label=f"val_norm={val_norm}")
axs[1].set_title("Data Setelah Normalisasi")
axs[1].set_xlabel("Index")
axs[1].set_ylabel("Nilai Normalisasi (0‚Äì1)")
axs[1].legend()
axs[1].set_xticks(range(len(x_norm)))
axs[1].set_xticklabels([f"x{i}" for i in range(len(x_norm))])

# Grafik batang data asli dengan titik hasil denormalisasi
axs[2].bar(range(len(x)), x, color='lightgreen')
axs[2].axhline(val_denorm, color='red', linestyle='--', label=f"val_denorm={val_denorm}")
axs[2].set_title("Denormalisasi (kembali ke skala asli)")
axs[2].set_xlabel("Index")
axs[2].set_ylabel("Nilai")
axs[2].legend()
axs[2].set_xticks(range(len(x)))
axs[2].set_xticklabels([f"x{i}" for i in range(len(x))])

plt.tight_layout()
plt.show()
```

üîé Penjelasan
-	Subplot 1: data asli [50, 100, 200, 300].
-	Subplot 2: data setelah normalisasi (0‚Äì1), dengan garis merah di 0.5.
-	Subplot 3: data asli lagi, dengan garis merah di nilai denormalisasi = 175 (hasil dari 0.5).
- üëâ Dengan ini, bisa melihat jelas bahwa angka 0.5 di dunia normalisasi kembali menjadi 175 di dunia asli.
