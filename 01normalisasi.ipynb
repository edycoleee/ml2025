# Normalisasi = “menyusutkan” semua nilai agar pas di antara 0–1 tanpa mengubah perbandingan antar-nilainya.
xnorm = (x - xmin) / (xmax-xmin)

```python
#contoh 1 konversi nilai [50, 100, 200, 300] ke 0 -1
import numpy as np
# Normalisasi
x = np.array([50, 100, 200, 300]) 
x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
print (x_norm)
#[0.  0.2 0.6 1. ]
#jadi 50 = 0, 100 = 0.2, 300 = 1
```

```python
#contoh 2 konversi nilai [ 50 150] ke 0 -1
x = np.array([50, 150])   # luas tanah
print(x)#[ 50 150]

x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
print (x_norm) #[0. 1.]
```

```python
#contoh 3 konversi nilai [300, 800] ke 0 -1
y = np.array([300, 800])  # harga rumah
y_norm = (y - np.min(y)) / (np.max(y) - np.min(y))
print (y_norm) #[0. 1.]
```

# Denormalisasi : yaitu proses mengubah skala 1 - 0 kembali ke data semula
y = (ynorm * (ymax-ymin)) + ymin

```python
#contoh 4 denormalisasi nilai 0.75
y_pred_norm = 0.75
y_pred = y_pred_norm * (np.max(y) - np.min(y)) + np.min(y)
print(f"{y_pred:.2f}")
```
